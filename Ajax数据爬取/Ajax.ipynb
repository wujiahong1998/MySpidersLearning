```python
from pyquery import PyQuery as pq
def parse_page(json):
    if json:
        items = json.get('data').get('cards')
        for item in items:
            weibo = {}
            weibo['id'] = item.get('mblog').get('id')
            weibo['text'] = pq(item.get('mblog').get('text')).text()
            weibo['attitudes'] = item.get('mblog').get('attitudes_count')
            weibo['comments'] = item.get('mblog').get('comments_count')
            weibo['reports'] = item.get('mblog').get('reposts_count')
            yield weibo
```
