```python
import requests
import re
```

设置请求头和url

```python
url = 'http://www.zmz2019.com/'
headers = {
    'User-Agent':'Opera/9.80 (Windows NT 6.0) Presto/2.12.388 Version/12.14'

}
```

获取整个网页文档

```python
r = requests.get(url,headers=headers)
r.encoding = 'utf-8'
r.text
```

保存网页文档

```python
text = r.text
```

编写正则表达式进行过滤

```python
regex = '<div class="box clearfix">.*?<ul>(.*?)</ul>.*?</div>'
result = re.search(regex,r.text,re.S)
result.group(1)
```

```python
regex = '<li>(.*?)</li>'
result = re.findall(regex,result.group(),re.S)
result
```

```python
regex_data = '<em>(.*?)</em>.*?<span>(\d+)</span>.*?<a href="(.*?)" title="(.*?)">.*?</a>'
data_list = []
for ri in result:
    r = re.search(regex_data,ri,re.S)
    print(r.groups())
    item = dict()
    item['type'] = r.group(1)
    item['index'] = r.group(2)
    item['link'] = 'http://www.zmz2019.com/' + r.group(3)
    item['name'] = r.group(4)
    if(item in data_list):
        continue
    data_list.append(item)
```

```python
for x in data_list:
    print(x)
```

```python
with open('top24-RR.md','w',encoding='utf-8') as f:
    f.write('# 24小时下载热门\n')
    for x in data_list:
        f.write("## " + str(x['index']) + x['name'] + '-' + x['type'] + '\n')
        f.write("link:" + x['link'] + '\n')
```
